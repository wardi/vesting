```python
from scipy.stats import truncnorm
days = 365 * 4
d = numpy.arange(days + 1, dtype=float)
```

```python
y_sin = ((-1j * 1j ** (2 * d / days)).imag + 1 ) / 2 * 100
y_sin = numpy.floor(y_sin)
y_cliff = d / days * 100
y_cliff = numpy.floor(y_cliff)
y_cliff[:365] = 0
y_norm10 = truncnorm(-5, 5).cdf((d / days - 0.5) * 10) * 100
y_norm10 = numpy.floor(y_norm10)
y_norm6 = truncnorm(-3, 3).cdf((d / days - 0.5) * 6) * 100
y_norm6 = numpy.floor(y_norm6)
y_norm4 = truncnorm(-2, 2).cdf((d / days - 0.5) * 4) * 100
y_norm4 = numpy.floor(y_norm4)
pylab.plot(d[:700], y_sin[:700], 'r',
           d[:700], y_cliff[:700],'g',
           d[:700], y_norm10[:700], 'b+',
           d[:700], y_norm6[:700], 'b',
           d[:700], y_norm4[:700], 'b--')
```
