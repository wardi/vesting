```python
from scipy.stats import truncnorm
days = 365 * 4
d = numpy.arange(days, dtype=float)[:700]
```

```python
y_sin = (-1j * 1j ** (2 * d / days)).imag / 2 + 0.5
y_cliff = d / days
y_cliff[:365] = 0
y_norm10 = truncnorm(-5, 5).cdf((d / days - 0.5) * 10)
y_norm6 = truncnorm(-3, 3).cdf((d / days - 0.5) * 6)
y_norm4 = truncnorm(-2, 2).cdf((d / days - 0.5) * 4)
pylab.plot(d, y_sin, 'r', d, y_cliff,'g', d, y_norm10, 'b+', d, y_norm6, 'b', d, y_norm4, 'b--')
```
